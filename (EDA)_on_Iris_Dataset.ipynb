```python
dataset=pd.read_csv("iris.csv")
dataset.head()
```

Petal Length vs. Petal Width

```python
sns.scatterplot(x="petal lengthcm ()",y="petal width (cm)",data=dataset,hue="species")
plt.show()
```

Sepal Length vs. Sepal Width

```python
sns.scatterplot(x="sepal length (cm)",y="sepal width (cm)",data=dataset,hue="species")
plt.show()
```

Petal Length of species

```python
sns.barplot(x="species",y="petal length (cm)",data=dataset,palette={'setosa':'green','versicolor':'orange','virginica':'blue'})
plt.show()
```
